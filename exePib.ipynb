```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import seaborn as sns
```

```python
df = pd.read_csv("pib.csv", sep=";")
df.head()
```

```python
# Dimensões do DataFrame
print("Dimensões do DataFrame:", df.shape)

# Tipos de dados das colunas
print("\nTipos de dados:")
print(df.dtypes)

# Verificar valores ausentes
print("\nValores ausentes:")
print(df.isnull().sum())

# Estatísticas descritivas
print("\nEstatísticas descritivas:")
print(df.describe())
```

```python
import matplotlib.pyplot as plt

# Histograma do PIB
plt.figure(figsize=(10, 5))
plt.hist(df['PIB'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuição do PIB')
plt.xlabel('PIB')
plt.ylabel('Frequência')
plt.show()

# Histograma do ValorEmpenho
plt.figure(figsize=(10, 5))
plt.hist(df['VALOREMPENHO'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Distribuição do ValorEmpenho')
plt.xlabel('ValorEmpenho')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['PIB'], df['VALOREMPENHO'], alpha=0.7)
plt.title('Relação entre PIB e ValorEmpenho')
plt.xlabel('PIB')
plt.ylabel('ValorEmpenho')
plt.grid(True)
plt.show()
```

```python
# Municípios com maior PIB
maior_pib = df.nlargest(5, 'PIB')
print("\nMunicípios com maior PIB:")
print(maior_pib[['MUNICIPIO', 'PIB']])

# Municípios com menor PIB
menor_pib = df.nsmallest(5, 'PIB')
print("\nMunicípios com menor PIB:")
print(menor_pib[['MUNICIPIO', 'PIB']])
```
