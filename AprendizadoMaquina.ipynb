```python
credito = pd.read_csv('credit.csv')
credito.shape
```

```python
previsor = credito.iloc[:,0:20].values
classe = credito.iloc[:,20].values
```

```python
labelencoder1 = LabelEncoder() 
previsor[:,0]= labelencoder1.fit_transform(previsor[:,0])
labelencoder2 = LabelEncoder() 
previsor[:,2]= labelencoder1.fit_transform(previsor[:,2])
labelencoder3 = LabelEncoder() 
previsor[:,3]= labelencoder1.fit_transform(previsor[:,3])
labelencoder4 = LabelEncoder() 
previsor[:,5]= labelencoder1.fit_transform(previsor[:,5])
labelencoder5 = LabelEncoder() 
previsor[:,6]= labelencoder1.fit_transform(previsor[:,6])
labelencoder6 = LabelEncoder() 
previsor[:,8]= labelencoder1.fit_transform(previsor[:,8])
labelencoder7 = LabelEncoder() 
previsor[:,9]= labelencoder1.fit_transform(previsor[:,9])
labelencoder8 = LabelEncoder() 
previsor[:,11]= labelencoder1.fit_transform(previsor[:,1])
labelencoder9 = LabelEncoder() 
previsor[:,13]= labelencoder1.fit_transform(previsor[:,13])
labelencoder10 = LabelEncoder() 
previsor[:,14]= labelencoder1.fit_transform(previsor[:,14])
labelencoder11 = LabelEncoder() 
previsor[:,16]= labelencoder1.fit_transform(previsor[:,16])
labelencoder12 = LabelEncoder() 
previsor[:,18]= labelencoder1.fit_transform(previsor[:,18])
labelencoder13 = LabelEncoder() 
previsor[:,19]= labelencoder1.fit_transform(previsor[:,19])
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsor, classe, test_size=0.3, random_state=0)
X_teste
```

```python
naive_bayes = GaussianNB()
naive_bayes.fit(X_treinamento, y_treinamento)
```

```python
previsoes = naive_bayes.predict(X_teste)
previsoes
```

```python
confusao = confusion_matrix(y_teste, previsoes)
confusao
```

```python
taxa = accuracy_score(y_teste, previsoes)
taxa_errp = 1 - taxa
taxa
```
